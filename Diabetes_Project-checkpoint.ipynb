PART I: DATA COLLECTION AND CLEANING

```python
import pandas as pd
```

```python
# read the csv into a DataFrame
df = pd.read_csv("diabetes_data_upload.csv")
df
```

```python
# replace the values in the DataFrame with binary values
df = df.replace('No', 0)
df = df.replace('Yes', 1)

###    df = df.replace({'Yes': 1, 'No': 0})

df = df.replace('Negative', 0)
df = df.replace('Positive', 1)

# consider Gender column as isMale
df = df.replace('Male', 1)
df = df.replace('Female', 0)

df
```

```python
# check for missing values
df.isnull().sum()
```

```python
# check the dtypes of the columns
df.dtypes
```

```python
replace = {'Gender': 'is_male'}

df = df.rename(columns = replace)

#replace columns to lowercase
df.columns.str.lower()
```

```python
df.columns = df.columns.str.lower()
df
```

```python
# export DataFrame to CSV
df.to_csv("diabetes_data_clean.csv", index = None)
```

### Questions:
1. Is obesity related to diabetes status?
2. Is age related to diabetes status?

```python
obesity_diabetes_crosstab = pd.crosstab(df['class'],df['obesity'])
obesity_diabetes_crosstab
```

```python
chi2_contingency(obesity_diabetes_crosstab)
```

```python
# Create a contingency table (counts of Obesity vs. Class)
obesity_vs_class = pd.crosstab(df['obesity'], df['class'])

# Normalize to get percentages instead of counts
obesity_vs_class_perc = obesity_vs_class.div(obesity_vs_class.sum(axis=1), axis=0) * 100

# Plot a stacked bar chart
obesity_vs_class_perc.plot(kind='bar', stacked=True, colormap='coolwarm', figsize=(6,4))

# Add labels and title
plt.title('Obesity vs. Diabetes Class', fontsize=14)
plt.xlabel('Obesity Status', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.legend(title='Diabetes Class', labels=['Non-Diabetic', 'Diabetic'])

# Show the plot
plt.show()
```

```python
is_male_diabetes_crosstab = pd.crosstab(df['class'],df['is_male'])
is_male_diabetes_crosstab
```

```python
chi2_contingency(is_male_diabetes_crosstab)
```

```python
# relationship between age and diabetic status
# age is in discete integers while diabetic status is categorical 

# Create the boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x='class', y='age', data=df, palette='deep')

# Add labels and title
plt.xlabel("Diabetes Class")
plt.ylabel("Age")
plt.title("Boxplot of Age vs Diabetes Class")

# Show the plot
plt.show()
```

```python
no_diabetes = df[df['class']==0]
no_diabetes['age'].median()
```

```python
diabetes = df[df['class']==1]
diabetes['age'].median()
```

```python
qqplot(df['age'], fit=True, line='s')
plt.show()

#distribution is normal if most points fall on the line
```

```python
# conduct z test of difference
ztest(diabetes['age'], no_diabetes['age'])

#conclusion == there is a relationship between age and diabetes
```
